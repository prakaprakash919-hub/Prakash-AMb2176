```python
#Arithmetic Operators
a = 15
b = 4

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b) 		# Divides and returns the integer part only
print("Modulus:", a % b)
print("Exponentiation:", a ** b)		# Raises the first number to the power of the second (ab)
print("Unary Plus:", +a)
print("Unary Minus:", -a)
```
