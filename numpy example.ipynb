```python
import numpy as np

# 1. Store Sales Data in a NumPy array
sales = np.array([1200, 1500, 1100, 1800, 1700, 1600, 1400])

# 2. Find Total Sales
total_sales = np.sum(sales)

# 3. Find Average Sales
avg_sales = np.mean(sales)

# 4. Identify specific performance days
highest = np.max(sales)
lowest = np.min(sales)

# 5. Filter data (Boolean Indexing)
high_performance_days = sales[sales > 1500]

# 6. Count days below average
low_days_count = np.sum(sales < avg_sales)

# --- Display Results ---
print(f"Weekly Sales Array: {sales}")
print("-" * 30)
print(f"Total Sales:        ₹{total_sales}")
print(f"Average Daily Sale: ₹{avg_sales:.2f}")
print(f"Highest Sale:       ₹{highest}")
print(f"Lowest Sale:        ₹{lowest}")
print(f"Sales above 1500:   {high_performance_days}")
print(f"Low Sales Days:     {low_days_count}")
```
