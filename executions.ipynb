```python
# Function to display all crop yields
def display_yields(yields_list):
    print("\nRecorded Crop Yields (kg):")
    for y in yields_list:
        print(y)


# Function to calculate average crop yield
def average_yield(yields_list):
    total = 0
    for y in yields_list:
        total = total + y
    avg = total / len(yields_list)
    return avg


# Function to find yields above a threshold value
def yields_above_threshold(yields_list, threshold):
    above = []
    for y in yields_list:
        if y > threshold:
            above.append(y)
    return above


# -------- MAIN PROGRAM --------

# Accept number of fields
fields = int(input("Enter number of fields: "))

crop_yields = []

# Collect crop yield for each field
for i in range(fields):
    yield_kg = float(input(f"Enter crop yield for field {i+1} (kg): "))
    crop_yields.append(yield_kg)

# Display all yields
display_yields(crop_yields)

# Calculate average yield
avg = average_yield(crop_yields)

# Find maximum and minimum yield
max_yield = max(crop_yields)
min_yield = min(crop_yields)

# Accept threshold value
threshold_value = float(input("\nEnter threshold yield value (kg): "))
above_threshold = yields_above_threshold(crop_yields, threshold_value)

# Display results
print("\n------ CROP YIELD SUMMARY ------")
print("Total number of fields:", fields)
print("Maximum yield:", max_yield, "kg")
print("Minimum yield:", min_yield, "kg")
print("Average yield:", avg, "kg")

print("\nYields above threshold:")
if len(above_threshold) > 0:
    for y in above_threshold:
        print(y)
else:
    print("No yields above the threshold value")
```

```python
# -------- DAILY EXPENSE MANAGEMENT --------

# Accept number of working days
days = int(input("Enter number of working days: "))

daily_expenses = []   # Nested list to store daily expense data
overall_expense = 0

# Collect expense details for each day
for i in range(days):
    print("\nEnter details for Day", i + 1)

    day_name = input("Enter day name: ")
    seed_cost = float(input("Enter seed cost: "))
    fertilizer_cost = float(input("Enter fertilizer cost: "))
    labour_cost = float(input("Enter labour cost: "))

    # Store data as a nested list
    daily_expenses.append([day_name, seed_cost, fertilizer_cost, labour_cost])


# -------- EXPENSE ANALYSIS --------

print("\n------ DAILY EXPENSE SUMMARY ------")

for day in daily_expenses:
    day_total = day[1] + day[2] + day[3]
    overall_expense = overall_expense + day_total

    print("Day:", day[0])
    print("Total expense:", day_total)
    print("--------------------------")

print("Overall farm expense:", overall_expense)


# -------- CROP INFORMATION MANAGEMENT --------

crops = []   # List to store crop records (nested tuples)

num_crops = int(input("\nEnter number of crops cultivated: "))

for i in range(num_crops):
    print("\nEnter crop details", i + 1)

    crop_name = input("Enter crop name: ")
    season = input("Enter season of cultivation: ")
    expected_yield = float(input("Enter expected yield (kg/ha): "))

    # Store crop details as a tuple (immutable)
    crop_record = (crop_name, season, expected_yield)
    crops.append(crop_record)


# -------- DISPLAY CROP DETAILS --------

print("\n------ CROP DETAILS ------")

for index, crop in enumerate(crops, start=1):
    print(index, ". Crop Name:", crop[0])
    print("   Season:", crop[1])
    print("   Expected Yield:", crop[2], "kg/ha")
```
