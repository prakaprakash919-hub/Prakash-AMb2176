```python
#Python - Update Tuples
#Change Tuple Values
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)

print(x)
```

```python
#Add Items(append())
thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.append("orange")
thistuple = tuple(y)
print(thistuple)
```

```python
#Add tuple to a tuple.
thistuple = ("apple", "banana", "cherry")
y = ("orange",)
thistuple += y

print(thistuple)
```

```python
#Remove Items
thistuple = ("apple","banana","cherry")
y = list(thistuple)
y.remove("apple")
thistuple = tuple(y)
print (thistuple)
```

```python
#Remove Items
thistuple = ("apple", "banana", "apple")
y = list(thistuple)
y.remove("apple")
thistuple = tuple(y)
print(thistuple)
```

```python
#The del keyword can delete the tuple completely
thistuple = ("apple", "banana", "cherry")
del thistuple
print(thistuple) #this will raise an error because the tuple no longer exists
```

```python
#Packing a tuple:
fruits = ("apple", "banana", "cherry")
print(fruits)
```

```python
#Unpacking a tuple:
fruits = ("apple", "banana", "cherry")

(green, yellow, red) = fruits

print(green)
print(yellow)
print(red)
```

```python
#Using Asterisk* Assign the rest of the values as a list called "red":
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")

(green, yellow, *red) = fruits

print(green)
print(yellow)
print(red)
```

```python
#Add a list of values the "tropic" variable:
fruits = ("apple", "mango", "papaya", "pineapple", "cherry")

(green, *tropic, red) = fruits

print(green)
print(tropic)
print(red)
```

```python
#Python - Loop Tuples
#Loop Through a Tuple for loop
thistuple = ("apple", "banana", "cherry")
for x in thistuple:
  print(x)
```

```python
#Loop Through the Index Numbers
thistuple = ("apple", "banana", "cherry")
for i in range(len(thistuple)):
  print(thistuple[i])
```

```python
#Print all items, using a while loop to go through all the index numbers:
thistuple = ("apple", "banana", "cherry")
i = 0
while i < len(thistuple):
  print(thistuple[i])
  i = i + 1
```

```python
#To join two or more tuples you can use the + operator:
tuple1 = ("a", "b" , "c")
tuple2 = (1, 2, 3)

tuple3 = tuple1 + tuple2
print(tuple3)
```

```python
#Multiply the fruits tuple by 2:
fruits = ("apple", "banana", "cherry")
mytuple = fruits * 2

print(mytuple)
```

```python
#SLICING TUPLES

tuple_data = (10, 20, 30, 40, 50)

print(tuple_data[:]) #prints all values
#this is same as print(tuple_data)
# and print(tuple_data[::])
# and print(tuple_data[::1])

print(tuple_data[::-1]) #prints all values in reverse

print(tuple_data[::2])
#O/P: [10,30,50]
```
