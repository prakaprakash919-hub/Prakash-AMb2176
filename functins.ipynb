```python
#C3 one argument is passed
def fahrenheit_to_celsius(fahrenheit):
  return (fahrenheit - 32) * 5 / 9

print(fahrenheit_to_celsius(77))
print(fahrenheit_to_celsius(95))
print(fahrenheit_to_celsius(50))
```

```python
def plant_info(plant_name):
  print(f"The {plant_name} requires at least 8 hours of sunlight per day")

plant_info("snake tree")
```

```python
#C5:one argument is passed,value is returned
def water_needs(plant_type):
  if plant_type == "corn":
    return 30
  elif plant_type == "wheat":
    return 20 #wheat
  else:
    return 25 #default water needs

water_needs("corn")
```

```python
# C6:Two arguments are passed, no value is returned

def irrigation_needed(moisture_level, rainfall):
    if moisture_level <= 40 and rainfall <= 10:
        print("Irrigation needed.")
    else:
        print("Irrigation not needed.")
irrigation_needed(30, 5)
```

```python
#A function with one argument:
def my_function(fname):
  print(fname + " Refsnes")

my_function("Emil")
my_function("Tobias")
my_function("Linus")
```

```python
#Parameters vs Arguments
def my_function(name): # name is a parameter
  print("Hello", name)

my_function("Emil") # "Emil" is an argument
```

```python
#C8:argument with default values
def my_function(name = "friend"):
  print("Hello", name)

my_function("Emil")
my_function("Tobias")
my_function()
my_function("Linus")
```

```python
#Keyword Arguments
#You can send arguments with the key = value syntax.
def my_function(animal, name):
  print("I have a", animal)
  print("My", animal + "'s name is", name)

my_function(name = "Buddy", animal = "dog")
```

```python
#Mixing Positional and Keyword Arguments
def my_function(animal, name, age):
  print("I have a", age, "year old", animal, "named", name)

my_function("dog", name = "Buddy", age = 5)
```
