#Matplotlib Scatter
With Pyplot, you can use the **scatter()** function to draw a scatter plot.

The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

```python
#A simple scatter plot:

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y, marker = "*")
plt.show()
```

The observation in the example above is the result of 13 cars passing by.

The X-axis shows how old the car is.

The Y-axis shows the speed of the car when it passes.

#ColorMap
The Matplotlib module has a number of available colormaps.

A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.

This colormap is called 'viridis' and as you can see it ranges from 0, which is a purple color, up to 100, which is a yellow color

You can specify the colormap with the keyword argument cmap with the value of the colormap, in this case 'viridis' which is one of the built-in colormaps available in Matplotlib.

```python
#Create a color array, and specify a colormap in the scatter plot:

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.show()
```

#Size
You can change the size of the dots with the s argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:

```python
#Set your own size for the markers:

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

case study:plot a bargraph analysing the no of runs scored in each over in T20 match
give diierent colours for max and min runs scored in over left overs defalt

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D

# 1. Prepare the Data (Sample T20 Match Overs and Runs)
overs = np.arange(1, 21)
runs = [8, 12, 5, 15, 10, 18, 6, 9, 11, 7, 14, 10, 5, 22, 16, 25, 12, 19, 21, 14]

# 2. Identify Max and Min Values
max_runs = max(runs)
min_runs = min(runs)

# 3. Define Colors
# Green for Max, Red for Min, Skyblue for others
colors = []
for r in runs:
    if r == max_runs:
        colors.append('green')
    elif r == min_runs:
        colors.append('red')
    else:
        colors.append('skyblue')

# 4. Create the Bar Plot
plt.figure(figsize=(12, 6))
plt.bar(overs, runs, color=colors, edgecolor='black')

# 5. Customize Labels and Title
plt.xlabel('Over Number', fontsize=12, fontweight='bold')
plt.ylabel('Runs Scored', fontsize=12, fontweight='bold')
plt.title('Runs Scored per Over in a T20 Match', fontsize=14, pad=15)
plt.xticks(overs) # Ensure every over is labeled on the X-axis
plt.grid(axis='y', linestyle='--', alpha=0.6)

# 6. Add Data Labels on top of each bar
for i, v in enumerate(runs):
    plt.text(overs[i], v + 0.3, str(v), ha='center', fontweight='bold')

# 7. Create a Custom Legend
legend_elements = [
    Line2D([0], [0], color='green', lw=6, label=f'Max Runs ({max_runs})'),
    Line2D([0], [0], color='red', lw=6, label=f'Min Runs ({min_runs})'),
    Line2D([0], [0], color='skyblue', lw=6, label='Default')
]
plt.legend(handles=legend_elements, loc='upper left')

# 8. Show Plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data for 20 overs in a T20 match
np.random.seed(42) # for reproducibility
overs = np.arange(1, 21)
runs_per_over = np.random.randint(low=2, high=25, size=20)

# Find the index of the over with maximum and minimum runs
max_runs_over_idx = np.argmax(runs_per_over)
min_runs_over_idx = np.argmin(runs_per_over)

# Create a list of colors for the bars
# Default color will be blue, max run over green, min run over red
bar_colors = ['skyblue'] * len(overs)
bar_colors[max_runs_over_idx] = 'green'
bar_colors[min_runs_over_idx] = 'red'

# Plotting the bar graph
plt.figure(figsize=(12, 6))
plt.barh(overs, runs_per_over, color=bar_colors)

# Add labels and title
plt.xlabel("Over Number", fontsize=12)
plt.ylabel("Runs Scored", fontsize=12)
plt.title("Runs Scored Per Over in a T20 Match (Max/Min Highlighted)", fontsize=14, fontweight='bold')

# Add legend for colors (optional, but good for clarity)
plt.legend(handles=[
    plt.Rectangle((0,0),1,1, color='green', label=f'Max Runs ({runs_per_over[max_runs_over_idx]} in Over {overs[max_runs_over_idx]})'),
    plt.Rectangle((0,0),1,1, color='red', label=f'Min Runs ({runs_per_over[min_runs_over_idx]} in Over {overs[min_runs_over_idx]})'),
    plt.Rectangle((0,0),1,1, color='skyblue', label='Other Overs')
])

plt.xticks(overs) # Ensure all over numbers are displayed on x-axis
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#Matplotlib Pie Charts
Creating Pie Charts
With Pyplot, you can use the pie() function to draw pie charts:



#Labels
Add labels to the pie chart with the labels parameter.

The labels parameter must be an array with one label for each wedge:

```python
#A simple pie chart:

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

start angle argument gives qudrant in pie chart by default start angle vlue is 0

#Shadow
Add a shadow to the pie chart by setting the shadows parameter to True:

```python
#Add a shadow:

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```
