#Matplotlib Adding Grid Lines
**Add Grid Lines to a Plot**
With Pyplot, you can use the grid() function to add grid lines to the plot.

note:the corresponding grids in diagonal with the coordinates 80,240) to (90,260) and 90,260 to (100,280) consists of same grid values

```python
#Add grid lines to the plot:

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

#Specify Which Grid Lines to Display
You can use the axis parameter in the grid() function to specify which grid lines to display.

Legal values are: 'x', 'y', and 'both'. Default value is 'both'.

```python
#Display only grid lines for the x-axis:axis = 'x

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'x')

plt.show()
```

#Matplotlib Subplot
**Display Multiple Plots**
With the subplot() function you can draw multiple plots in one figure:

```python
#Draw 2 plots:

import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

#The subplot() Function
The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

the subplots are plotted side by side
suppose we want to the 2 subplot on top of eachother use PLT.SUBPLOT(2,1,2)

```python
#Draw 2 plots on top of each other:

import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and a 5x4 grid of subplots
fig, axs = plt.subplots(5, 4, figsize=(15, 12))

# Flatten the array of axes for easy iteration
axs = axs.flatten()

# Data for plotting (example data)
x = np.array([0, 1, 2, 3])
y_base = np.array([3, 8, 1, 10])

# Populate each subplot
for i in range(20):
    ax = axs[i]
    # Generate slightly different data for each subplot
    y = y_base + i * 0.5
    ax.plot(x, y)
    ax.set_title(f'Plot {i+1}')
    ax.tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False,
                   labelbottom=False, labelleft=False)

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```

#Matplotlib Basic Plotting
11.1 Explanation with Definition
Matplotlib is the foundational plotting library in Python, providing comprehensive tools for creating static, animated, and interactive visualizations. It offers two interfaces: pyplot (MATLAB-style) and object-oriented (OO). Understanding basic plots is essential for exploratory data analysis and communicating ML results.

#Package and Method Explanations
plt.plot(x, y): Creates line plot with optional markers and styles
plt.scatter(x, y): Creates scatter plot with customizable point size and colors
plt.bar(x, height): Creates vertical bar chart
plt.hist(data, bins): Creates histogram showing data distribution
plt.xlabel() / plt.ylabel(): Sets axis labels
plt.title(): Sets plot title
plt.legend(): Displays legend for labeled plot elements
plt.grid(): Adds grid lines for easier reading
plt.figure(figsize): Creates new figure with specified size in inches
plt.tight_layout(): Automatically adjusts subplot parameters for clean layout
plt.savefig(): Saves figure to file (png, jpg, pdf, svg)
plt.twinx(): Creates second y-axis sharing same x-axis

```python
import matplotlib.pyplot as plt
import numpy as np

# Agricultural data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rainfall = np.array([45, 52, 68, 120, 180, 220, 195, 160, 110, 75, 58, 40])
temperature = np.array([18, 20, 24, 28, 32, 35, 34, 33, 30, 26, 22, 19])
wheat_yield = np.array([2.1, 2.3, 2.8, 4.2, 5.8, 6.5, 5.9, 5.2, 3.8, 2.9, 2.4, 2.0])

# Example 1: Line Plot - Rainfall Over Time
plt.figure(figsize=(10, 5))
plt.plot(months, rainfall, marker='o', linewidth=2,
         color='#1f77b4', label='Rainfall')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Rainfall (mm)', fontsize=12)
plt.title('Monthly Rainfall Pattern', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
# plt.savefig('rainfall_pattern.png', dpi=300)
print('Figure 1: Line plot created - Monthly Rainfall Pattern')

# Example 2: Multiple Lines - Temperature and Rainfall
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(months, rainfall, marker='s', color='blue', label='Rainfall')
ax1.set_xlabel('Month', fontsize=12)
ax1.set_ylabel('Rainfall (mm)', color='blue', fontsize=12)
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(months, temperature, marker='^', color='red', label='Temperature')
ax2.set_ylabel('Temperature (°C)', color='red', fontsize=12)
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Rainfall and Temperature Trends', fontsize=14, fontweight='bold')
fig.tight_layout()
print('Figure 2: Dual-axis plot created - Rainfall and Temperature')

# Example 3: Bar Chart - Crop Yields by Region
regions = ['North', 'South', 'East', 'West', 'Central']
yields = [180, 228, 416, 164, 245]
colors_list = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc']

plt.figure(figsize=(10, 6))
bars = plt.bar(regions, yields, color=colors_list, edgecolor='black', linewidth=1.5)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Yield (tons)', fontsize=12)
plt.title('Crop Yield by Region', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.0f}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

print('Figure 3: Bar chart created - Crop Yield by Region')

# Example 4: Scatter Plot - Rainfall vs Yield Correlation
plt.figure(figsize=(10, 6))
plt.scatter(rainfall, wheat_yield, s=100, c=temperature,
            cmap='YlOrRd', edgecolors='black', linewidth=1, alpha=0.7)
plt.colorbar(label='Temperature (°C)')
plt.xlabel('Rainfall (mm)', fontsize=12)
plt.ylabel('Wheat Yield (tons/ha)', fontsize=12)
plt.title('Rainfall vs Yield Correlation (colored by temperature)',
          fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
print('Figure 4: Scatter plot created - Rainfall vs Yield')

# Example 5: Histogram - Yield Distribution
plt.figure(figsize=(10, 6))
plt.hist(wheat_yield, bins=8, color='green', alpha=0.7,
         edgecolor='black', linewidth=1.5)
plt.xlabel('Yield (tons/ha)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Wheat Yields', fontsize=14, fontweight='bold')
plt.axvline(wheat_yield.mean(), color='red', linestyle='--',
            linewidth=2, label=f'Mean: {wheat_yield.mean():.2f}')
plt.legend()
plt.grid(axis='y', alpha=0.3)
print('Figure 5: Histogram created - Yield Distribution')

print('\nAll plots generated successfully!')
print('Note: Use plt.show() to display plots interactively')
```
