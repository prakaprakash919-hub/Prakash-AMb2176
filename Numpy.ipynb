### `range()` Function

`range()` is a built-in Python function that generates a sequence of integers. It's primarily used for looping a specific number of times and for creating sequences of integers when you need a basic iterable.

```python
print('--- Using range() ---')

# Example with integers
sequence_range = range(0, 5) # Generates numbers from 0 up to (but not including) 5
print(f"range(0, 5): {list(sequence_range)}")
print(f"Type of range(0, 5): {type(sequence_range)}")

# range() does not support floating-point numbers
try:
    range_float = range(0, 1, 0.5)
except TypeError as e:
    print(f"\nError when using floats with range(): {e}")
```

### `np.arange()` Function

`np.arange()` is a NumPy function that also generates sequences of numbers. However, it's more versatile, capable of generating sequences with floating-point numbers and returning a NumPy array, which is highly optimized for numerical operations.

```python
import numpy as np

print('\n--- Using np.arange() ---')

# Example with integers
sequence_arange_int = np.arange(0, 5)
print(f"np.arange(0, 5): {sequence_arange_int}")
print(f"Type of np.arange(0, 5): {type(sequence_arange_int)}")

# Example with floating-point numbers
sequence_arange_float = np.arange(0.0, 1.0, 0.2)
print(f"np.arange(0.0, 1.0, 0.2): {sequence_arange_float}")
print(f"Type of np.arange(0.0, 1.0, 0.2): {type(sequence_arange_float)}")
```

### Key Differences:

1.  **Return Type**: `range()` returns a `range` object (an iterator that generates numbers on the fly), while `np.arange()` returns a NumPy `ndarray`.
2.  **Data Types**: `range()` only works with integers. `np.arange()` can handle both integers and floating-point numbers.
3.  **Memory Usage**: `range()` is memory-efficient because it generates numbers as needed (lazy evaluation). `np.arange()` creates an array in memory containing all the numbers immediately.
4.  **Functionality**: `np.arange()` provides the full power of NumPy arrays, allowing for vectorized operations, advanced indexing, and integration with other NumPy functions, which `range()` does not.

#Basic Array Operations
In numpy, arrays allow a wide range of operations which can be performed on a particular array or a combination of Arrays. These operation include some basic Mathematical operation as well as Unary and Binary operations.

```python
import numpy as np

# Defining Array 1
a = np.array([[1, 2],
              [3, 4]])

# Defining Array 2
b = np.array([[4, 3],
              [2, 1]])

# Adding 1 to every element
print ("Adding 1 to every element:", a + 1)

# Subtracting 2 from each element
print ("\nSubtracting 2 from each element:", b - 2)

# sum of array elements
# Performing Unary operations
print ("\nSum of all array elements: ", a.sum())

# Adding two arrays
# Performing Binary operations
print ("\nArray sum:\n", a + b)
```

```python
import numpy as np

# Integer datatype
x = np.array([1, 2])
print(x.dtype)

# Float datatype
x = np.array([1.0, 2.0])
print(x.dtype)

# Forced Datatype
x = np.array([1.0, 2.1], dtype = np.int64)
print(x.dtype)
```

#Math Operations on DataType array
In Numpy arrays, basic mathematical operations are performed element-wise on the array. These operations are applied both as operator overloads and as functions. Many useful functions are provided in Numpy for performing computations on Arrays such as sum: for addition of Array elements, T: for Transpose of elements, etc.

```python
#sum
import numpy as np

# First Array
arr1 = np.array([[4, 7], [2, 6]],
                 dtype = np.float64)

# Second Array
arr2 = np.array([[3, 6], [2, 8]],
                 dtype = np.float64)

# Addition of two Arrays
Sum = np.add(arr1, arr2)
print(Sum)

# Addition of all Array elements
Sum1 = np.sum(arr1)
print(Sum1)

# Square root of Array
Sqrt = np.sqrt(arr1)
print(Sqrt)

# Transpose of Array
Trans_arr = arr1.T
print(Trans_arr)
```

```python
yields = np.array([2500, 2800, 3100, 2900, 2700])

farm_grid = np.array([
    [2500, 2600, 2700],
    [2800, 2900, 3000],
    [2400, 2500, 2600]
])

# === ARRAY OPERATIONS ===
print("\n=== Array Operations ===")

# Math on entire arrays (vectorized!)
yields_tons = yields / 1000
print(f"Yields in tons: {yields_tons}")

# Bonus calculation
bonus = np.where(yields > 2800, 500, 0)
print(f"Bonus per plot: {bonus}")

# Statistics
print(f"\nTotal yield: {np.sum(yields)} kg")
print(f"Average: {np.mean(yields):.0f} kg")
print(f"Std Dev: {np.std(yields):.0f} kg")
print(f"Max: {np.max(yields)} kg, Min: {np.min(yields)} kg")

# Boolean filtering
high_yield_plots = yields[yields > 2700]
print(f"High yield plots (>2700): {high_yield_plots}")

low_yield_plots = yields[yields < 2700]
print(f"Low yield plots (<2700): {low_yield_plots}")
```
