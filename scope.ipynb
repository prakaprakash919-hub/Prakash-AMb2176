```python
#Local Scope
def myfunc():
  x = 300
  print(x)

myfunc()
```

```python
#Global Scope
x = 300

def myfunc():
  x=300
  print(x)

myfunc()

print(x)
```

```python

x = 300

def myfunc():
  x = 200
  print(x)

myfunc()

print(x)
```

```python
#Lambda Functions:1 arguments
x = lambda a : a + 10
print(x(5))
```

```python
#Multiply argument a with argument b and return the result:2 arguments
x = lambda a, b : a * b
print(x(5, 6))
```

```python
#lambda with 3 arguments
x = lambda a, b, c : a + b + c
print(x(5, 6, 2))
```

```python
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)

print(mydoubler(11))
```

```python
def myfunc(n,p):
  return lambda a,b:a*n+p*b

mydoubler = myfunc(2,3)

print(mydoubler(11,22))
```

```python
words = ["apple", "pie", "banana", "cherry"]
sorted_words = sorted(words, key=lambda x: len(x))
print(sorted_words)
```
